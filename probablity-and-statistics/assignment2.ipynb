```python
df = df.iloc[:, 1:]

stats = []
labels = []

for col in df.columns:
    data = df[col].dropna()

    q0 = data.min()
    q1 = data.quantile(0.25)
    q2 = data.quantile(0.5)
    q3 = data.quantile(0.75)
    q4 = data.max()

    stats.append({
        "med": q2,
        "q1": q1,
        "q3": q3,
        "whislo": q0,
        "whishi": q4,
        "fliers": []
    })

    labels.append(col)

fig, ax = plt.subplots()
ax.bxp(stats, showfliers=False)
ax.set_xticklabels(labels)
ax.set_title("Boxplot of All Subjects")
ax.set_ylabel("Marks")

plt.show()
```
