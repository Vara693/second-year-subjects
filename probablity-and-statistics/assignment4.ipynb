```python
def factorial(n):
    res = 1;
    for i in range(1, n+1):
        res *= i
    return res;
```

```python
def binomialCoeff(n, r):
    return factorial(n)/(factorial(n-r)*factorial(r))
```

```python
def binomialProbablity(n, r, p):
    coeff = binomialCoeff(n, r)
    return coeff*(p**r)*((1-p)**(n-r))
```
