```python
import random
import matplotlib.pyplot as plt
```

```python
t = 10000
count_b = 0
count_b_a = 0
x = []
y = []
```

```python
for i in range(0, t):
    num = random.randint(1,6)
    if num>3:
        count_b +=1
        if num%2==0:
            count_b_a +=1
    if count_b>0:
        x.append(i)
        y.append(count_b_a/count_b)

print(count_b_a/count_b)
print(len(x))

plt.plot(x, y)
plt.show()
```
