```python
import matplotlib.pyplot as plt
import math
import pandas as pd
from scipy.stats import norm
```

```python
def mean(arr):
    n = len(arr)
    ans = 0
    for num in arr:
        ans += num
    return ans/n

def variance(arr):
    n= len(arr)
    ans = 0
    x = mean(arr)
    for num in arr:
        ans += (num-x)**2
    return ans/n
```

```python
df = pd.read_csv("mark.csv")
```

```python
def normal(mean, var, x):
    a = 1/(((2*math.pi)**0.5) * var)
    b = 1/(2*(var))
    c = (x - mean)**2

    return a*(math.exp(-b*c))
```

```python
marks = df["Marks"]
```

```python
x_ = mean(marks)
var = variance(marks)
ndf = []
```

```python
for i in marks:
    ndf.append(normal(x_, var, i))
```

```python
plt.scatter(marks, ndf, color="blue", label="Data points")

y_mean = mean(ndf)

num = sum((marks[i] - x_) * (ndf[i] - y_mean) for i in range(len(marks)))
den = sum((marks[i] - x_) ** 2 for i in range(len(marks)))
slope = num / den
intercept = y_mean - slope * x_

y_trend = [slope * xi + intercept for xi in marks]

plt.plot(marks, y_trend, color="red", label="Trend line")
```
