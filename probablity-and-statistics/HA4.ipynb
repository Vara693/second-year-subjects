```python
import numpy as np
from scipy import stats

before = np.array([45, 50, 55, 60, 48, 52, 49])
after = np.array([50, 55, 60, 65, 52, 58, 54])

d = after - before

mean_d = np.mean(d)
std_d = np.std(d, ddof=1)
n = len(d)

t_stat = mean_d / (std_d / np.sqrt(n))

# Degrees of freedom
df = n - 1

t_critical = stats.t.ppf(1 - 0.025, df)

print("t-statistic:", t_stat)
print("t-critical:", t_critical)

if abs(t_stat) > t_critical:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

```python
import numpy as np
from scipy import stats

machine_A = np.array([102, 105, 98, 100, 110, 108, 104])
machine_B = np.array([95, 97, 92, 96, 94, 93, 91])

mean_A, mean_B = np.mean(machine_A), np.mean(machine_B)
std_A, std_B = np.std(machine_A, ddof=1), np.std(machine_B, ddof=1)

n1, n2 = len(machine_A), len(machine_B)

t_stat = (mean_A - mean_B) / np.sqrt((std_A**2 / n1) + (std_B**2 / n2))

# Degrees of freedom
df = n1 + n2 - 2

t_critical = stats.t.ppf(1 - 0.025, df)

print("t-statistic:", t_stat)
print("t-critical:", t_critical)

if abs(t_stat) > t_critical:
    print("Reject H0")
else:
    print("Fail to reject H0")
```
