```python
import random
import matplotlib.pyplot as plt
```

```python
t = 10000
c_b = 0
c_a_b = 0
```

```python
x = []
y = []
```

```python
for i in range(0,t):
  die = random.randint(1,6)
  if die > 3:
    c_b += 1
    if die % 2 ==0:
      c_a_b += 1
  if c_b>0:
    x.append(i)
    y.append(c_a_b/c_b)
```

```python
cp = c_a_b/c_b
```

```python
cp
```

```python
plt.plot(x,y)
plt.show()
```
