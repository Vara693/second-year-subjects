```python
from scipy import stats

before = [45, 50, 55, 60, 48, 52, 49]
after = [50, 55, 60, 65, 52, 58, 54]

t_stat, p_value = stats.ttest_rel(before, after)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
from scipy import stats

machine_A = [102, 105, 98, 100, 110, 108, 104]
machine_B = [95, 97, 92, 96, 94, 93, 91]

t_stat, p_value = stats.ttest_ind(machine_A, machine_B)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```
