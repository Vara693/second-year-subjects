```python
def median(arr):
    n= len(arr)
    # med = 0
    if (n%2==0):
        return (arr[n//2] + arr[n//2 - 1])/2
    else:
        return arr[n//2]
    # return med

def mean(arr):
    n = len(arr)
    ans = 0
    for num in arr:
        ans += num
    return ans/n
```

```python
median([1,2,3,4,5,6])
```

```python
median(df["Sub 2"])
```

```python
df.median(axis=1)
```

```python
def variance(arr):
    n= len(arr)
    ans = 0
    x = mean(arr)
    for num in arr:
        ans += (num-x)**2
    return ans/n
```

```python
arr = [1,2,3,4,5]
variance(arr)
```

```python
nums = [2, 9, 8, 6, 4]
variance(nums)
```

```python
def standardDeviation(arr):
    return variance(arr)**0.5
```

```python
nums = [2, 9, 8, 6, 4]
standardDeviation(nums)
```
